```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#read in data
econ = pd.read_csv('data/EconInd.csv')
ma_fin = pd.read_csv('data/MA_fin.csv') #financial info
ma_ww = pd.read_csv('data/MA_WW.csv') #worldwide KPIs
v_fin = pd.read_csv('data/V_fin.csv') #financial info
v_ww = pd.read_csv('data/V_WW.csv') #worldwide KPIs

#convert data to consistent format
econ = econ.rename(columns={'Unnamed: 0':'tag_time'}).set_index('tag_time')
ma_fin = ma_fin.rename(columns={'Report Date':'tag_time'}).set_index('tag_time')
ma_ww = ma_ww.rename(columns={'Report Date':'tag_time'}).set_index('tag_time')
v_fin = v_fin.rename(columns={'Report Date':'tag_time'}).set_index('tag_time')
v_ww = v_ww.rename(columns={'Report Date':'tag_time'}).set_index('tag_time')

#merge financial and KPI dfs
ma = pd.merge(ma_fin,ma_ww, on='tag_time')
visa = pd.merge(v_fin,v_ww, on='tag_time')

#add dfs to dictionary
data = {'econ': econ, 'ma': ma, 'visa': visa}
```

```python
#EDA
plt.figure()
plt.plot(ma['Revenue'])
plt.plot(visa['Revenue'])
```
